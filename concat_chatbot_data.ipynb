```python
import pandas as pd
```

```python
chatbot_A = pd.read_csv('./data/chatbot_dataset/chatbot.csv')
chatbot_A
```

```python
chatbot_A = chatbot_A.drop(columns='Unnamed: 0')
chatbot_A
```

```python
chatbot_B = pd.read_csv('./data/chatbot_dataset/chatbot_data.csv')
chatbot_B
```

```python
chatbot_B.columns=['SPEAKER', 'SENTENCE', 'MAIN']
chatbot_B
```

```python
chatbot_B['SPEAKER'] = chatbot_B['SPEAKER'].replace('s', '점원')
```

```python
chatbot_B['SPEAKER'] = chatbot_B['SPEAKER'].replace('c', '고객')
```

```python
chatbot_B
```

```python
chatbot = pd.concat([chatbot_A, chatbot_B])
chatbot
```

```python
chatbot.to_csv('./data/chatbot_dataset/chatbot_all.csv')
```
